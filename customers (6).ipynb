# 1. Описание данных
## 1.1. Информация о массиве данных
Датасет содержит информацию о покупателях некоторого магазина. Массив данных был взят с сайта kaggle: https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis/data

В массиве представлены следующие переменные:
- ID: Пользовательский идентификатор клиента
- Year_Birth: год рождения Клиента
- Education: уровень образования клиента
- Marital_Status: Семейное положение клиента
- Income: годовой доход семьи клиента
- Kidhome: Количество детей в семье клиента
- Teenhome: Количество подростков в семье клиента

- Dt_Customer: Дата регистрации клиента в компании
- Recency: Количество дней с момента последней покупки клиента
- Complain: 1, если клиент жаловался в течение последних 2 лет, 0 в противном случае

- MntWines: Сумма, потраченная на вино за последние 2 года
- MntFruits: Сумма, потраченная на фрукты за последние 2 года
- MntMeatProducts: Сумма, потраченная на мясо за последние 2 года
- MntFishProducts: Сумма, потраченная на рыбу за последние 2 года
- MntSweetProducts: Сумма, потраченная на сладости за последние 2 года
- MntGoldProds: Сумма, потраченная на золото за последние 2 года

- NumDealsPurchases: Количество покупок, совершенных со скидкой
- AcceptedCmp1: 1, если клиент принял предложение в рамках 1-й кампании, в противном случае 0
- AcceptedCmp2: 1, если клиент принял предложение во время 2-й кампании, в противном случае 0
- AcceptedCmp3: 1, если клиент принял предложение в рамках третьей кампании, 0 в противном случае
- AcceptedCmp4: 1, если клиент принял предложение в 4-й кампании, в противном случае 0
- AcceptedCmp5: 1, если клиент принял предложение в 5-й кампании, в противном случае 0
- Response: 1, если клиент принял предложение в последней кампании (промо-акции), в противном случае 0

- NumWebPurchases: Количество покупок, совершенных через веб-сайт компании
- NumCatalogPurchases: Количество покупок, совершенных с использованием каталога
- NumStorePurchases: Количество покупок, совершенных непосредственно в магазинах
- NumWebVisitsMonth: Количество посещений веб-сайта компании за последний месяц



## 1.2. Описание цели исследования
Цель: предсказать, участвует ли покупатель в маркетинговых компаниях (промо-акциях). 

**Исследовательский вопрос:** какие характеристики покупателей увеличивают их склонность участвовать в промо-акциях

В качестве целевой переменной будет участие в последней компании. 

Участие в последней кампании будет предсказываться через персональные данные о покупателе (год рождения, уровень образования, семейное положение, уровень дохода, наличие детей, подростков - Year_Birth, Education, Marital_Status, Income, Kidhome, Teenhome), данные о взаимодействиях с магазином (дата регистрации, количество дней с момента последней покупки, наличие жалоб - Dt_Customer, Recency, Complain), характер трат клиента (сумма, потраченная на вино, фрукты, мясо, рыбу, сладости и золото за последние 2 года - MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds), покупки по скидке (NumDealsPurchases), участие в прошлых компаниях (AcceptedCmp1, AcceptedCmp2, AcceptedCmp3, AcceptedCmp4, AcceptedCmp5), а также способы покупки (через сайт, через каталог, в магазине - NumWebPurchases, NumCatalogPurchases, NumStorePurchases), а так же количество посещений сайта за последний  месяц (NumWebVisitsMonth)

## Используемые модели

В данной работе будет применяться knn классификация, логистическая регрессия, случайный лес и CatBoost, так как они подходят для бинарной классификации:

- Логистическая регрессия хорошо работает с линейно разделимыми данными и может быть интерпретирована, что делает ее привлекательным выбором для объяснения важности признаков
- Случайный лес подходит для задач бинарной классификации, так как сочетает в себе множество решающих деревьев, что позволяет справляться с разнообразием данных и избегать переобучения. Он хорошо работает с большими объемами данных, шумом и различными типами признаков
- CatBoost хорошо подходит для задач бинарной классификации, особенно если в данных присутствуют категориальные признаки

Ящики с усами

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Объединяем данные из двух датасетов
Response_1['Response'] = 1
Response_0['Response'] = 0
combined_data_2 = pd.concat([Response_1, Response_0])

# Фильтруем переменные для анализа, начинающиеся с 'Mnt'
Mnt_vars = [col for col in combined_data_2.columns if col.startswith('Mnt')]

# Построение графиков с усами для каждой переменной Mnt_vars по 3 в строке
num_subplots = len(Mnt_vars)
rows = 2
cols = num_subplots // 2 + num_subplots % 2
plt.figure(figsize=(18, 12))

# Задаем цвета для значений 0 и 1
colors = {0:'#7CF56E', 1:'#7C6EF5'}

for i, mnt_var in enumerate(Mnt_vars):
    plt.subplot(rows, cols, i+1)
    sns.boxplot(data=combined_data_2, x='Response', y=mnt_var, hue='Response', palette=colors)
    plt.title(f'График с усами для {mnt_var}')

plt.tight_layout()
plt.show()
```

- С ростом доходов склонность попасть в категорию участников кампании растет

Попробобуем оптимизировать параметры для улучшения качества модели

```python
accuracy_scores = []
f1_scores = []

for n in range(3, 16):
    # Создаем классификатор с текущим значением n_neighbors
    knn = KNeighborsClassifier(n_neighbors=n)
    
    # Обучаем модель и делаем прогнозы
    knn.fit(X_train_norm, y_train)
    y_pred_knn = knn.predict(X_test_norm)
    
    # Вычисляем точность и F1-меру
    accuracy = accuracy_score(y_test, y_pred_knn)
    f1 = f1_score(y_test, y_pred_knn)
    
    # Добавляем значения в списки
    accuracy_scores.append(accuracy)
    f1_scores.append(f1)

    print(f"n_neighbors = {n}: Accuracy = {accuracy}, F1 = {f1}")

# Выводим лучшее значение метрик
best_accuracy_idx = np.argmax(accuracy_scores)
best_f1_idx = np.argmax(f1_scores)

print(f"\nЛучший результат точности: n_neighbors = {best_accuracy_idx + 3}, Accuracy = {accuracy_scores[best_accuracy_idx]}")
print(f"Лучший результат F1-меры: n_neighbors = {best_f1_idx + 3}, F1 = {f1_scores[best_f1_idx]}")
```

Оптимальный результат - 3 соседа. Оценим данную модель

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_norm, y_train)
y_pred_knn = knn.predict(X_test_norm)
p_pred_knn = knn.predict_proba(X_test_norm) 
p_pred_knn = p_pred_knn[:, 1]  # P(y = 1 | X)
```

Получим метрики качества этой модели

```python
print(classification_report(y_test, y_pred_knn))
```

Даная модель лучше, чем модель с характеристиками по умолчанию, но f1-score для класса 1 все равно маленькая (0,41). По другим показателям (accuracy, f1-score для класса 0) данная модель не лучше нулевой модели

```python
from sklearn import linear_model #линейные модели
#Создаем объект класса логистическая регрессия
log_reg = linear_model.LogisticRegression(
    random_state=1, #генератор случайных чисел
)
#Обучаем модель, минимизируя logloss
log_reg.fit(X_train_norm, y_train)
```

Выведем метрики качества для обучающей и тестовой выборок:

```python
from sklearn import metrics #метрики
#Делаем предсказание для тренировочной выборки
y_train_pred = log_reg.predict(X_train_norm)
#Вывод отчет о метриках классификации
print(metrics.classification_report(y_train, y_train_pred))
#Делаем предсказание для тестовой выборки
y_test_pred = log_reg.predict(X_test_norm)
#Вывод отчет о метриках классификации
print(metrics.classification_report(y_test, y_test_pred))
```

Видим, что по accuracy и f1-score модель лучше базовой и knn, по f1-score для  класса 1, модель уже лучше, чем knn (0,55 против 0,41), но снова показатель нельзя назвать очень хорошим

Попробуем изменить порог отсечения для улучшения качества классификации

```python
#Нас интересует только вероятность класса (второй столбец)
y_test_proba_pred = log_reg.predict_proba(X_test_norm)[:, 1]
#Для удобства завернем numpy-массив в pandas Series
y_test_proba_pred = pd.Series(y_test_proba_pred)
#Создадим списки, в которых будем хранить значения метрик 
recall_scores = []
precision_scores = []
f1_scores = []
#Сгенерируем набор вероятностных порогов в диапазоне от 0.1 до 1
thresholds = np.arange(0.1, 1, 0.05)
#В цикле будем перебирать сгенерированные пороги
for threshold in thresholds:
    #В противном случае - к классу 0
    y_test_pred = y_test_proba_pred.apply(lambda x: 1 if x>threshold else 0)
    #Считаем метрики и добавляем их в списки
    recall_scores.append(metrics.recall_score(y_test, y_test_pred))
    precision_scores.append(metrics.precision_score(y_test, y_test_pred))
    f1_scores.append(metrics.f1_score(y_test, y_test_pred))

#Визуализируем метрики при различных threshold
fig, ax = plt.subplots(figsize=(10, 4)) #фигура + координатная плоскость
#Строим линейный график зависимости recall от threshold
ax.plot(thresholds, recall_scores, label='Recall')
#Строим линейный график зависимости precision от threshold
ax.plot(thresholds, precision_scores, label='Precision')

#Строим линейный график зависимости F1 от threshold
ax.plot(thresholds, f1_scores, label='F1-score')
#Даем графику название и подписи осям
ax.set_title('Recall/Precision dependence on the threshold')
ax.set_xlabel('Probability threshold')
ax.set_ylabel('Score')
ax.legend();
```

Находим значение оптимального порога

```python
max_f1_index = np.argmax(f1_scores)
# Найдем значение threshold для этой точки
optimal_threshold = thresholds[max_f1_index]

print("Optimal Threshold value:", optimal_threshold)
```

Получаем метрики качества модели с учетом этого порога 

```python
#Задаем оптимальный порог отсеченяи
threshold_opt = optimal_threshold
y_test_pred_opt = y_test_proba_pred.apply(lambda x: 1 if x > threshold_opt else 0)
#Считаем метрики
print(metrics.classification_report(y_test, y_test_pred_opt))
```

accuracy и f1-score  для класса 0 сократились, но они не хуже, чем в базовой модели, f1-score для класса 1 выросла до 0,6, что лучше, чем во всех моделях, которые были раньше, но такой показатель тоже не идеал, хотя для настолько несбалансированных классов не плохо

Найдем оптимальные пераметры:

```python
from sklearn.model_selection import GridSearchCV
from sklearn import ensemble, metrics

# Задаем сетку параметров, которые мы хотим перебрать
param_grid = {
    'n_estimators': [100, 300, 500], #количество деревьев в лесу
    'max_depth': [10, 15, 20], # максимальная глубина каждого дерева
    'criterion': ['gini', 'entropy'], #критерий для разделения узлов дерева
    'min_samples_leaf': [5, 10, 15] #минимальное число объектов, необходимое для того, чтобы узел дерева мог бы расщепиться
}

# Создаем объект GridSearchCV с моделью случайного леса и сеткой параметров
grid_search = GridSearchCV(estimator=ensemble.RandomForestClassifier(random_state=42, class_weight='balanced'),
                           param_grid=param_grid,
                           scoring='f1',
                           cv=3,
                           n_jobs=-1)

# Выполняем поиск лучших параметров
grid_search.fit(X_train, y_train)


# Выводим лучшие параметры и значение метрики
print("Лучшие параметры:", grid_search.best_params_)
print('Train: {:.2f}'.format(grid_search.best_score_))

# Получаем предсказания с лучшей моделью и выводим значение метрики на тестовой выборке
best_rf = grid_search.best_estimator_
y_test_pred = best_rf.predict(X_test)
print('Лучший результат на тестовой выборке (f1 score): {:.2f}'.format(metrics.f1_score(y_test, y_test_pred)))
```
